```python
def first_unique_char(s):
    freq = {}

    # Count frequency of each character
    for ch in s:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    # Find the first character with frequency 1
    for i in range(len(s)):   # Iterate through the string
        if freq[s[i]] == 1:      # If the character is unique
            return i          # Return the index of the first unique character

    return -1

# Example usage
s = "examination"
print(first_unique_char(s))  # Output: 0
```

```python
# Assignment 3 - 4/8/2025


#  1. The Lost and Found Box (Duplicate Item Detection)
# Story:
# You’re in charge of a Lost & Found box in your college. Every time someone reports a lost item, you check if it’s already in the box.
# Problem: How do you avoid checking every item one by one (which is slow)?
# Solution using Hashing:
# You keep a set (hash table) of item names like {"watch", "pen", "bottle"}.
# Every time someone reports a new item, you check the hash table:
# •	If it's there: "Yes! Found it!"
# •	If not: Add to the set and store the item.
# ➡ Real use: This is how computers check for duplicate files, logins, or data super fast.

class lost_item:
    def __init__(self):
        self.hash_table = {}

    def report(self, item):
        key = hash(item) 

        if key in self.hash_table:
            print("Yes! Found it!")
        else:
            print("Not found. Adding to the box.")
            self.hash_table[key] = item

box = lost_item()
box.report("pen")
box.report("bottle")
box.report("pen")  
box.report("watch")
```

```python

#  2. Secret Diary Lock (Password Hashing)
# Story:
# You write in your diary and lock it with a password. But you don’t want to write the real password on the lock—it’s unsafe.
# Problem: How do you store the lock's code without exposing your real password?
# Solution using Hashing:
# You hash the password like this:
# •	Real password: "sunshine123"
# •	Hashed: "5ae9f8abc..."
# When you enter your password next time, the diary hashes it again and compares the hashed version.
# ➡ Real use: This is how websites store your passwords safely without knowing your actual password.

class secretdiary:
    def __init__(self, password):
        self.hashed_password = self.fake_hash(password)

    def fake_hash(self, text):
        return sum((i + 1) * len(c) for i, c in enumerate(text))

    def unlock(self, attempt):
        if self.fake_hash(attempt) == self.hashed_password:
            print("Diary Unlocked! ")
        else:
            print("Wrong password! ")

diary = secretdiary("sunshine123")
diary.unlock("sunshine123") 
diary.unlock("sunshine") 
```

```python
# 5. Same Essay? (Plagiarism Detection)
# Story:
# You’re checking student essays. You suspect two students copied each other.
# Problem: Reading both essays manually is slow and boring.
# Solution using Hashing:
# You convert each essay into a hash value. If the hash is same → The content is likely same.
# ➡ Real use: Used by apps like Turnitin or Google Docs to check copy-paste content.

class essaychecker:
    def __init__(self):
        pass

    def h_essay(self, text):
        words = text.lower().split()
        return sum(len(word) for word in words) + len(set(words))

    def similarity(self, essay1, essay2):
        hash1 = self.h_essay(essay1)
        hash2 = self.h_essay(essay2)

        if hash1 == hash2:
            print("Copied")
        else:
            print("Not Copied")

checker = essaychecker()

essay_a = "Razi is not a good boy"
essay_b = "Razi is not a good boy"
essay_c = "Razeen also not good boy"

checker.similarity(essay_a, essay_b)
checker.similarity(essay_a, essay_c)  
```

```python
# Q2. Evaluate Postfix Expression (Stack)
# Scenario:
# You are building a calculator that evaluates expressions like:
# "5 3 + 2 *" → (5 + 3) * 2 = 16
# Task:
# Write a function evaluate_postfix(expr: str) that returns the result.
# Example:
# Input: "5 3 + 2 *"
# Output: 16
# Hint: Use a stack. Push numbers, pop 2 on operator, compute and push result.

def evaluate_postfix(expr: str):
    stack = []
    tokens = expr.split()

    for token in tokens:
        if token.isdigit():
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()
            
            if token == '+':
                result = a + b
            elif token == '-':
                result = a - b
            elif token == '*':
                result = a * b
            elif token == '/':
                result = a // b 
            else:
                return "Invalid operator"
            
            stack.append(result) 

    return stack.pop() if stack else "Invalid expression"

print(evaluate_postfix("5 3 + 2 *"))  
```

```python
#  Q3. Menu Combinations for Calories (Backtracking)
# Scenario:
# A restaurant wants to find all combinations of dishes that sum up to exactly 700 calories for diet plans. Each dish has a calorie value, and dishes can be repeated.
# Task:
# Given a list of dish calorie values, and a target, find all combinations where dishes sum to target calories.
# Example:
# dishes = [200, 250, 150]
# target = 700
# Tags: Backtracking, combination sum, menu planning    


def find(dishes, target):
    result = []

    def backtrack(start, path, total):
        if total == target:
            result.append(path.copy())
            return
        if total > target:
            return

        for i in range(start, len(dishes)):
            path.append(dishes[i])
            backtrack(i, path, total + dishes[i])  # i, not i+1 (dishes can repeat)
            path.pop()

    backtrack(0, [], 0)
    return result

# Example usage
dishes = [200, 250, 150]
target = 700

combinations = find(dishes, target)
for i, combo in enumerate(combinations, 1):
    print(f"Combo {i}: {combo}")
```

```python
# Q4. Phone Number to Words (Recursion + Backtracking)
# Scenario:
# You're building a contact suggestion system. Given a digit sequence like on a dialpad (2 → abc, 3 → def, etc.), generate all valid letter combinations.
# Task:
# Input: "23"
# Output: ["ad", "ae", "af", "bd", "be", "bf", "cd", "ce", "cf"]
# Tags: Recursion, backtracking, predictive text



def phone(digits):
    if not digits:
        return []

    digit= {
        "2": "abc", "3": "def",
        "4": "ghi", "5": "jkl",
        "6": "mno", "7": "pqrs",
        "8": "tuv", "9": "wxyz"
    }

    result = []

    def backtrack(index, path):
        if index == len(digits):
            result.append("".join(path))
            return

        for letter in digit[digits[index]]:
            path.append(letter)
            backtrack(index + 1, path)
            path.pop()

    backtrack(0, [])
    return result

# Test
print(phone("23"))
```

```python
# Assignment 5  - 7/8/2025

# 1. Organizational Hierarchy (Tree View)
# Scenario:
# A company’s reporting structure:
#           CEO
#          /   \
#     Manager1 Manager2
#      /             \
#   Employee1     Employee2
# Why Binary Tree?
# •	Each person can report to one manager (parent), and may have up to 2 direct subordinates.
# Use-case:
# •	Traversals (level order, in-order) can be used to generate org charts or send emails level-wise.



class Node():
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None


root = Node("CEO")
root.left = Node("Manager1")
root.right = Node("Manager2")
root.left.left = Node("Employee1")
root.right.right = Node("Employee2")


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")
        
def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print("    " * level, node.val)
        print_tree(node.left, level + 1)


print_tree(root)
print("\n")
print("Inorder Traversal")
inorder(root)
print("\n")
print("preorder Traversal")
preorder(root)
print("\n")
print("postorder Traversal")
postorder( root)
print("\n")
```

```python
# 2. Tournament Bracket System
# Scenario:
# Any knockout-style tournament:
#        Final
#       /     \
#   Semi1     Semi2
#    /  \     /  \
#  Q1  Q2   Q3  Q4
# Why Binary Tree?
# •	Each match has two players (left/right).
# •	The winner moves up as the parent node (next match).
# Use-case:
# •	Winner determination, path to final, etc.



class Node():
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


q1 = Node("Alice")
q2 = Node("Bob")
q3 = Node("Raju")
q4 = Node("Radha")

semi1 = Node("Alice")
semi1.left = q1
semi1.right = q2

semi2 = Node("Raju")
semi2.left = q3
semi2.right = q4

final = Node("Raju")
final.left = semi1
final.right = semi2


def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")

def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)

print_tree(final)
print("\n")
print ("The Winner is: Raju ")
print("\n")
print("Inorder Traversal")
inorder(final)
print("\n")
print("preorder Traversal")
preorder(final)
print("\n")
print("postorder Traversal")
postorder(final)
print("\n")
```

```python
# 3. Expression Evaluation in Compilers (Expression Tree)
# Scenario:
# Evaluate math expression:
# (3 + 5) * (2 - 1)
# Tree:
#        *
#      /   \
#     +     -
#    / \   / \
#   3  5  2  1
# Why Binary Tree?
# •	Each operator has 2 operands → left/right children
# Use-case:
# •	Post-order traversal helps evaluate expression
# •	In-order traversal reconstructs it




class Node():
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


root = Node("*")
root.left = Node("+")
root.right = Node("-")

root.left.left = Node("3")
root.left.right = Node("5")
root.right.left = Node("2")
root.right.right = Node("1")


def inorder(root):
    if root:
        if root.left or root.right:
            print("(", end="")
        inorder(root.left)
        print(root.val, end=" ")
        inorder(root.right)
        if root.left or root.right:
            print(")", end="")

def preorder(root):
    if root:
        print(root.val, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end=" ")


def evaluate(root):
    if not root.left and not root.right:
        return int(root.val)
    l = evaluate(root.left)
    r = evaluate(root.right)
    if root.val == '+': return l + r
    if root.val == '-': return l - r
    if root.val == '*': return l * r
    if root.val == '/': return l // r 


def print_tree(node, level=0):
    if node:
        print_tree(node.right, level + 1)
        print("    " * level + node.val)
        print_tree(node.left, level + 1)


print_tree(root)
print("\n")
print("Inorder Traversal")
inorder(root)
print("\n")
print("preorder Traversal")
preorder(root)
print("\n")
print("postorder Traversal")
postorder( root)
print("\n")
```
